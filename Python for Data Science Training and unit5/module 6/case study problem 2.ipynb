```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../datasets/cars.csv')
data.head()
```

# Question 1

```python
x = data['cyl'].unique()
y = data['cyl'].value_counts()
```

```python
plt.bar(x,y)
plt.xlabel('number of cylinders')
plt.ylabel('Frequency of cars')
```
