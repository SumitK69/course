```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Question 1

```python
data = pd.read_csv('../datasets/customer_churn.csv')
data.head()
```

```python
x = data['Contract'].unique()
y = data['Contract'].value_counts()
```

```python
plt.bar(x,y,color = 'orange')
plt.xlabel('contract type of customer')
plt.ylabel('Count')
plt.title("Distribution of Contract")
plt.show()
```

# Question 2

```python
plt.hist(data['MonthlyCharges'],color = 'forestgreen')
plt.xlabel('Monthly charges incurred')
plt.ylabel('Count')
plt.title("Distribution of Monthly charges")
plt.show()
```

# Question 3

```python
x = data['tenure']
y = data['TotalCharges']
```

```python
plt.scatter(x,y,color = 'indigo')
plt.xlabel('tenure of the customer')
plt.ylabel('total charge incurred')
plt.title("total charge vs tenure")
plt.show()
```

# Question 4


```python
data.boxplot(column='MonthlyCharges',by = 'PaymentMethod',color= 'olive')
plt.xlabel('payment method of customer')
plt.ylabel('monthly charge incurred')
plt.title("monthly charge vs payment method")
plt.show()
```
