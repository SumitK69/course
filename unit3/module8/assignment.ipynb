```python
import pandas as pd
import numpy as np
from keras import models
from sklearn.datasets import load_breast_cancer
from keras.layers import Dense
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
```

```python
data_df = pd.DataFrame(data = data.data,
                       columns = data.feature_names)
data_df
```

```python
data_df['outcome'] = data.target
data_df
```

```python
x = data_df.iloc[:,0:-1]
y = data_df.iloc[:,-1:]
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.33)
import livelossplot
plot_losses = livelossplot.PlotLossesKeras()
%matplotlib inline
BATCH_SIZE = 10
EPOCHS = 100
```

```python
model = models.Sequential([
        layers.Dense(16, activation="relu", name="layer1"),
        layers.Dense(16, activation="softmax", name="layer2")])
model.compile(optimizer="adam",loss="SparseCategoricalCrossentropy",metrics = ['accuracy'])
model.fit(X_train,Y_train,batch_size = BATCH_SIZE,epochs = EPOCHS,callbacks = [plot_losses],verbose = 1,validation_data = (X_test,Y_test))
```

```python
score = model.evaluate(X_test,Y_test,verbose = 0)
print('test loss: ',score[0])
print('test accuracy: ',score[1])
```
