```python
import pandas as pd
from keras import models
from keras.layers import Dense
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers
```

```python
data = pd.read_csv("../datasets/diabetes.csv")
```

```python
data.head()
```

```python
x = data.iloc[:,:3]
y = data.iloc[:,-1:]
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.30)
```

```python
import livelossplot
plot_losses = livelossplot.PlotLossesKeras()
%matplotlib inline
BATCH_SIZE = 10
EPOCHS = 100
```

```python
model = models.Sequential([
        layers.Dense(5, activation="relu", name="layer1"),
        layers.Dense(5, activation="relu", name="layer2"),
        layers.Dense(6, name="layer3")])
model.compile(optimizer="adam",loss="SparseCategoricalCrossentropy",metrics = ['accuracy'])
model.fit(X_train,Y_train,batch_size = BATCH_SIZE,epochs = EPOCHS,callbacks = [plot_losses],verbose = 1,validation_data = (X_test,Y_test))
```

```python
score = model.evaluate(X_test,Y_test,verbose = 0)
print('test loss: ',score[0])
print('test accuracy: ',score[1])
```
