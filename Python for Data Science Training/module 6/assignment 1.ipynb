```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('../datasets/cars.csv')
cars.head()
```

```python
x = cars['model']
y = cars['hp']
```

```python
plt.plot(x,y)
plt.xticks(rotation = 90)
plt.title("model names vs horse power")
plt.xlabel("models of the cars")
plt.ylabel("horse power of cars")
plt.show()
```
